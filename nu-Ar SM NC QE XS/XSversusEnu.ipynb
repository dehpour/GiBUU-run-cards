```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Input these from GiBUU 2023 by run card.job for each Enu and then read XS from neutrino_absorption_cross_section_QE.dat

Enu = np.array([0.01,0.1,1,10,100])
sigma = [0.28129E-04,0.71661E-02,0.15208,0.14972,0.11170]
sigma = np.multiply(sigma,1e-38)
```

```python
plt.figure(figsize=(8, 6))

plt.plot(Enu,np.divide(sigma,Enu), label='GiBUU 2023')

plt.xscale('log')
#plt.yscale('log')

plt.title(r'$\nu_{\mu}-^{40}_{18}\ {\rm Ar}$ SM NC QE X.S.')
plt.xlabel(r'$E_{\nu}\ [{\rm GeV}]$')
plt.ylabel(r'$\sigma$ per nucleon $[{\rm cm}^2]/E_{\nu}$')
plt.legend()
plt.show()
```
